```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

Load the data and visualise the columns

```python
melbourne_file_path = '../input/melbourne-housing-snapshot/melb_data.csv'
melbourne_data = pd.read_csv(melbourne_file_path)
melbourne_data.columns
```

drop all the empty cells

```python
melbourne_data = melbourne_data.dropna(axis=0)
```

define target

```python
y = melbourne_data.Price
```

defining features

```python
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
x = melbourne_data[melbourne_features]
```

```python
x.describe()
```

```python
x.head()
```

selecting Decision tree model for fitting

```python
train_x, val_x, train_y, val_y = train_test_split(x,y, random_state=1)
model = DecisionTreeRegressor(random_state=1)
model.fit(train_x,train_y)
```

```python
print('Making predictions for the following houses: ')
print(train_x.head())
print('The predicted prices are: ')
predictions = model.predict(train_x)
print(predictions)
```

```python
mea = mean_absolute_error(train_y, predictions)
print(mea)
```

```python
val_predictions = model.predict(val_x)
val_mae = mean_absolute_error(val_y, val_predictions)
print(val_mae)
```

Overfitting and underfitting validation

```python
def get_mae(max_leaf_nodes, train_x, train_y, val_x, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes= max_leaf_nodes, random_state = 1)
    model.fit(train_x, train_y)
    prediction = model.predict(val_x)
    mae = mean_absolute_error(val_y, prediction)
    return mae
```

```python
for max_leaf_node in [5,50,500,5000]:
    mae = get_mae(max_leaf_node, train_x, train_y, val_x, val_y)
    print(max_leaf_node , mae)
```

finding the best tree size

```python
candidate_max_leaf_nodes = [5,50,100,200,250,500,5000]
scores = {max_leaf: get_mae(max_leaf,train_x, train_y, val_x, val_y) for max_leaf in candidate_max_leaf_nodes}
print(scores)
best_tree_size = min(scores, key=scores.get)
print(best_tree_size)
```

final Decision Tree model

```python
final_model = DecisionTreeRegressor(max_leaf_nodes = 250, random_state = 1)
final_model.fit(x, y)
final_model.predict(x)
```
